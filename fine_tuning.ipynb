```python
test_prompts = [
    "오늘 일이 잘 안 풀린다. 어떻게 해야 하지?",
    "공연 전 떨려서 죽을 것 같아. 한마디 해줘.",
    "요즘 사람들한테 잘 못 보이는 것 같아 속상해.",
    "갑자기 무언가를 도전하고 싶은데 겁이 나. 용기 좀 줘.",
    "형, 친구랑 싸웠어. 화해하려면 어떻게 말해야 할까?",
    "늦잠 자서 모든 일정 망쳤다 — 위로해줘.",
    "갑자기 돈키호테 저자를 묻는 팬이 있는데 어케 답할까?",
    "외국에서 한국말에 대해 자랑하려고 해. 한 문장 만들어줘.",
    "무대에서 크게 실수했어. 다시 설 수 있을까?",
    "인생에서 작은 행복을 찾고 싶어. 추천 좀 해줘.",
    "요리 초보야. 간단하지만 인상 깊은 한마디 레시피 알려줘.",
    "삶이 막막할 때 가장 먼저 해야 할 한 가지는?"
]
```

```python
for idx, prompt in enumerate(test_prompts, start=1):
    messages = [
        {"role": "system", "content": "무한도전의 길 느낌(철학적이면서 허당인)으로 답변하세요."},
        {"role": "user", "content": prompt}
    ]

    try:
        completion = client.chat.completions.create(
            model=fine_tuned_model_id,
            messages=messages,
            temperature=0.3,
            max_tokens=200
        )

        response_text = completion.choices[0].message.content.strip()

    except Exception as e:
        print("Error:", e)

    print(f"Q{idx}: {prompt}")
    print(f"A{idx}: {response_text}\n")
```
